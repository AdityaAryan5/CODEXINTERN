```python
import numpy as np

def input_matrix(n):
    print(f"\nEnter elements for Matrix {n} (rows separated by newlines, elements by space):")
    rows = int(input(f"Enter number of rows for Matrix {n}: "))
    cols = int(input(f"Enter number of columns for Matrix {n}: "))
    matrix = []
    for i in range(rows):
        row = list(map(float, input(f"Row {i+1}: ").strip().split()))
        if len(row) != cols:
            print("Column count mismatch. Try again.")
            return input_matrix(n)
        matrix.append(row)
    return np.array(matrix)

def display_matrix(mat, name="Matrix"):
    print(f"\n{name}:\n{mat}")

def matrix_operations():
    print(" Welcome to the Matrix Operations Tool")

    A = input_matrix("A")
    B = input_matrix("B")

    while True:
        print("\nSelect an operation to perform:")
        print("1. Addition (A + B)")
        print("2. Subtraction (A - B)")
        print("3. Multiplication (A × B)")
        print("4. Transpose of A")
        print("5. Transpose of B")
        print("6. Determinant of A")
        print("7. Determinant of B")
        print("8. Exit")

        choice = input("Enter your choice (1-8): ")

        if choice == "1":
            try:
                result = A + B
                display_matrix(result, "A + B")
            except:
                print("Error: Addition requires matrices of same dimensions.")

        elif choice == "2":
            try:
                result = A - B
                display_matrix(result, "A - B")
            except:
                print("Error: Subtraction requires matrices of same dimensions.")

        elif choice == "3":
            try:
                result = A @ B
                display_matrix(result, "A × B")
            except:
                print("Error: Matrix multiplication requires A's columns = B's rows.")

        elif choice == "4":
            display_matrix(A.T, "Transpose of A")

        elif choice == "5":
            display_matrix(B.T, "Transpose of B")

        elif choice == "6":
            try:
                if A.shape[0] != A.shape[1]:
                    raise ValueError
                det = np.linalg.det(A)
                print(f"Determinant of A: {det:.4f}")
            except:
                print("Error: Determinant can only be calculated for square matrices.")

        elif choice == "7":
            try:
                if B.shape[0] != B.shape[1]:
                    raise ValueError
                det = np.linalg.det(B)
                print(f"Determinant of B: {det:.4f}")
            except:
                print("Error: Determinant can only be calculated for square matrices.")

        elif choice == "8":
            print("Exiting Matrix Operations Tool. Goodbye!")
            break

        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    matrix_operations()
```
